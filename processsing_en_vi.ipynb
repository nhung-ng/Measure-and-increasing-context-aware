```python
#segment
!pip3 install vncorenlp
from vncorenlp import VnCoreNLP
rsegmenter = VnCoreNLP("/content/gdrive/MyDrive/KLTN/vncorenlp/VnCoreNLP-1.1.1.jar", annotators="wseg", max_heap_size='-Xmx500m')
```

```python
with open("/content/gdrive/MyDrive/KLTN/iwslt2015/PhoMT/tokenization/train/train.vi","r", encoding = "utf-8") as f:
    train_vi = f.read().split("\n")
    f.close()
```

```python
pre_train_vi = []
for line in train_vi:
    line = rsegmenter.tokenize(line)
    new_line = str()
    for sen in line:
        sen = " ".join(sen)
        new_line += str(sen)
    pre_train_vi.append(new_line)
```

```python
with open("/content/gdrive/MyDrive/KLTN/iwslt2015/PhoMT/en-vi/tmp/train.vi",'a', encoding="utf-8") as f:
    for i in range(len(pre_train_vi)):
        f.writelines(pre_train_vi[i])
        f.writelines("\n")
    f.close()
```
